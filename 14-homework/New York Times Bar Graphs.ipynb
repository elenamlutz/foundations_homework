```python
df1['Social Spending'] = df1['Social Spending'].replace('%', '', regex=True)
```

```python
df1['total'] = df1['Country'] + ' ' + df1['Social Spending']
```

```python
df1['Social Spending'] = df1['Social Spending'].astype(float)
```

```python
ax = df1.sort_values(by='Social Spending').plot(x='total', 
                                           y='Social Spending', 
                                           kind='barh',
                                           color='#CFDBEB',
                                           stacked=True,
                                           width=0.8,
                                           figsize=(3,10),
                                           title="Public social spending as a share of G.D.P.")
plt.savefig('nytbar1.pdf')
```

```python
ax = df.sort_values(by='year').plot(x='year', 
                               y='lead', 
                               kind='barh',
                               width=0.8,
                               figsize=(7,5),
                               title="2018 Doesn’t Look Quite Like Other Election Cycles")

for i, bar in enumerate(ax.patches):
    if i < 1.6:
        value = int(bar.get_width())
        x_pos = bar.get_width()
        y_pos = i
        ax.text(x_pos, y_pos, value, va='center')
    else:
        value = int(bar.get_width())
        x_pos = bar.get_width() - 1
        y_pos = i
        ax.text(x_pos, y_pos, value, va='center', ha='right', color='white', fontweight='bold')

plt.savefig('nytbar2.pdf')        
```
