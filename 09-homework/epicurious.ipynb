```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.epicurious.com/search/cucumbers')
doc = BeautifulSoup(response.text)
```

```python
all_texts = doc.find_all(class_='recipe-content-card')
rows = []
for text in all_texts:
    row={}
    try:   
        row['Category'] = text.find('strong', class_='tag').text.strip()
    
    except:
        print("Failed to find category")
    
    try:
        row['Title'] = text.find('a').text.strip()
    
    except:
        print("Failed to find title")

    try:
        row['Summary'] = text.find(class_='dek').text.strip()
    
    except:
        print("Failed to find summary")
    
    try:
        row['Rating'] = text.find(class_='rating').text.strip()
    
    except:
        print('Failed to find rating')
    
    try:
        row['Would Make Again?'] = text.find(class_='make-again-percentage').text.strip()
    
    except:
        print('Failed to find')
        
    try:
        row['URL'] = text.find('a')['href']
    
    except:
        print('Failed to find URL')
    
    rows.append(row)
df = pd.DataFrame(rows, columns=['Category', 'Title', 'Summary', 'Rating', 'Would_Make_Again', 'URL'])
```

```python
df['Rating'] = df.Rating.str.extract("(.*)/")
```

```python
rows = []

for page_num in range(0,16):
    url = f'https://www.epicurious.com/search/cucumbers?page={page_num}'
    
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    
    all_texts = doc.find_all(class_='recipe-content-card')

    for text in all_texts:
        row={}
      
        try:   
            row['Category'] = text.find('strong', class_='tag').text.strip()

        except:
            print("Failed to find category")

        try:
            row['Title'] = text.find('a').text.strip()

        except:
            print("Failed to find title")

        try:
            row['Summary'] = text.find(class_='dek').text.strip()

        except:
            print("Failed to find summary")

        try:
            row['Rating'] = text.find(class_='rating').text.strip()

        except:
            print('Failed to find rating')

        try:
            row['Would Make Again?'] = text.find(class_='make-again-percentage').text.strip()

        except:
            print('Failed to find')

        try:
            row['URL'] = text.find('a')['href']

        except:
            print('Failed to find URL')


        rows.append(row)

df = pd.DataFrame(rows)
```

```python
df['Rating'] = df.Rating.str.extract("(.*)/")
```
