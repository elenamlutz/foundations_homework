```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('http://kcna.kp')
```

```python
items = driver.find_elements_by_class_name('titlebet')
rows = []
for item in items:
    row = {}
    row['headline'] = item.text
    row['onclick'] = item.get_attribute('onclick')
    row['ID'] = item.get_attribute('onclick')[16:25]
    print(row)
    rows.append(row)
```

```python
import pandas as pd
df = pd.DataFrame(rows)
df
```

```python
df.to_csv('north_korean_news', index=False)
```
